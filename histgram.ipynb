```python
%matplotlib notebook
import matplotlib.pyplot as plt
import numpy
```

```python
import os
from os import listdir
from os.path import isfile, join
prjdir="/home/meip-users/Desktop/proen/proen_VC"
tgt="kozuka"
onlyfiles = [f for f in listdir(prjdir+"/data/f0/"+tgt+"/") if isfile(join(prjdir+"/data/f0/"+tgt+"/",f))]
print(onlyfiles[4])
```

```python

v =[]
#for fn in onlyfiles:
# fn ="VOICEACTRESS100_084.f0"
for fn in onlyfiles:
    with open(prjdir+"/data/f0/"+tgt+"/"+fn) as f:
        for row in f:
            #print(row.split()[0])
            x=float(row.split()[0])
            if x != 0:
                v.append(x)

plt.hist(v, 200)
plt.show()
```
